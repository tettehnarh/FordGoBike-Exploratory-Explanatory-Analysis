## Cleaning Data

```python
# Make copies of original data before cleaning effort
gobike_clean = df_gobike.copy()
```

### Quality Issues

### Issue #1: wrong datatypes in columns: start_time, end_time, bike_share_for_all_trip, bike_id, start_station_id, end_station_id and user_type

#### Define
<ul>
    <li>Change <code>start_time</code> and <code>end_time</code> to datetime datatype</li>
    <li>Change <code>bike_share_for_all_trip</code> to bool datatype</li>
    <li>Change <code>bike_id</code>, <code>start_station_id</code> and <code>end_station_id</code> to string datatype</li>
    <li>Change <code>user_type</code> to category datatype</li>
</ul>

#### Code

```python
#Converting datatype to datetime
gobike_clean.start_time = pd.to_datetime(gobike_clean.start_time)
gobike_clean.end_time = pd.to_datetime(gobike_clean.end_time)

#Converting datatype to bool
gobike_clean.bike_share_for_all_trip = gobike_clean.bike_share_for_all_trip.map({'No': False, 'Yes': True})

#Converting datatype to bool
gobike_clean.bike_id = gobike_clean.bike_id.astype(str)
gobike_clean.start_station_id = gobike_clean.start_station_id.astype(str)
gobike_clean.end_station_id = gobike_clean.end_station_id.astype(str)

#Converting datatype to category
gobike_clean.user_type = gobike_clean.user_type.astype('category')
```

#### Test

```python
gobike_clean.info()
```

### Issue #2:  null values in start_station_id, start_station_name, end_station_id, end_station_name, member_birth_year and member_gender

#### Define
Drop rows with null values

#### Code

```python
gobike_clean.dropna(axis=0, inplace=True)
```

#### Test 

```python
gobike_clean.isnull().sum()
```

### Housecleaning for Exploratory Analysis
#### Define 

<ul>
    <li>Create a new column <code>duration_min</code> from converting <code>duration_sec</code> to minutes for better representation and drop <code>duration_sec</code></li>
    <li>Create a new column <code>age</code>, based on bikers age as at 2019 with respect to <code>member_birth_year</code> and drop <code>member_birth_year</code> column</li>
    <li>Drop columns not required in analysis - <code>start_station_id</code>, <code>start_station_latitude</code>, <code>start_station_longitude</code>, <code>end_station_latitude</code>, <code>end_station_longitude</code>, <code>bike_id</code></li>
    <li>Check for outliers</li>
</ul>

#### Code

```python
#Convert duration in seconds to minutes
gobike_clean['duration_sec'] = round((gobike_clean['duration_sec']/60),2)
gobike_clean.rename(columns={'duration_sec':'duration_min'},inplace=True)
```

```python
#Create a column age, based on bikers age as at 2019 with respect to member_birth_year
gobike_clean['age'] = 2019 - gobike_clean['member_birth_year']
gobike_clean['age'] = gobike_clean['age'].astype(int)
gobike_clean.drop('member_birth_year',axis=1)
```

```python
#Drop uneccesary columns
gobike_clean = gobike_clean.drop(['start_station_id','start_station_name','end_station_id','member_birth_year',
                                  'bike_id','start_station_latitude','start_station_longitude','end_station_name',
                                  'end_station_latitude','end_station_longitude'], axis=1)
```

```python
#check for outliers
gobike_clean.describe(percentiles = [.25, .50, .75, .95, .99]).round(2)
```

Some outliers exist in our dataset. Age from 18 to 64 represent 99% of our users. A ride duration of 1,409 minutes also seems outrageous. It's logical to remove users and duration above the 99% percentile.

```python
# Cut outliers at 99th percentile for 'age' and 'duration'
gobike_clean = gobike_clean[gobike_clean['age']<=gobike_clean['age'].quantile(0.99)]
gobike_clean = gobike_clean[gobike_clean['duration_min']<=gobike_clean['duration_min'].quantile(0.99)]
```

## Univariate Exploration

### Question
What is the distribution of the trip duration?

```python
binsize = 0.2
bins = np.arange(0, gobike_clean['duration_min'].max()+binsize, binsize)

plt.figure(figsize=[10, 7])
plt.hist(data = gobike_clean, x = 'duration_min',bins=bins)
plt.title('Distribution of trip durations(linear scaled)', fontsize=15)
plt.xlabel('Trip duration in minutes', fontsize=12);
plt.ylabel('Count of trips', fontsize=12)
plt.show();
```

The distribution is largely right-skewed. I would apply log on the x-axis and limiting scope to 30 minutes.

```python
#Transforming to logscale and limiting x-axis to 30 minutes
plt.figure(figsize=[10, 7])
binsize = 0.2
bins = np.arange(0, gobike_clean['duration_min'].max()+binsize, binsize)

ticks = [1, 2, 3, 4, 5, 7, 10, 15, 20, 30, 50, 85]
labels = ['{}'.format(v) for v in ticks]
plt.hist(data = gobike_clean, x = 'duration_min', bins=bins);
plt.xscale('log')
plt.xticks(ticks, labels);
plt.xlim((1,30));
plt.title('Distribution of trips durations(log scaled)', fontsize=15)
plt.xlabel('Trip duration in minutes', fontsize=12)
plt.ylabel('Count of trips', fontsize=12)
plt.show()
```

<strong>Observation:</strong> The distribution is normal and most of the trips are done within 4-8 minutes

```python
#Function to create barplots
def myBarPlot(df, xVar, yVar, hueVar, palette=None, order=None, hue_order=None):
    '''
    Function to generate a bar plot based on the arguements provided
    palette, order, hue_order are set to default values
    inputs:
    ------
    xVar = x-axis variable
    yVar = y-axis variable
    hueVar = hue variable
    
    output:
    -------
    Generates Barplot with appropriate labels and values at x and y axis
    '''
    #Bar plot dimensions
    plt.figure(figsize=[14,6])

    #Plot graph
    sb.barplot(data=df, x=xVar, y=yVar, hue=hueVar, palette=palette, order=order, hue_order=hue_order);
    #Replacing _ with spaces for better representation
    xVar = xVar.replace("_"," ")
    yVar = yVar.replace("_"," ")
    hueVar = hueVar.replace("_"," ")
    #Add title
    plt.title(f'''Average of {yVar} by {xVar} and {hueVar}'''.title(),fontsize=14, weight="bold")
    #Add x-axis label
    plt.xlabel(xVar.title(), fontsize=10, weight='bold')
    #Add y-axis label
    plt.ylabel(f'Average of {yVar}'.title(),fontsize=10, weight="bold")
    #Despine
    sb.despine()
```

```python
#set order of values
order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
gobike_clean['start_day'] = gobike_clean['start_time'].dt.day_name()
#Call plot function
myBarPlot(gobike_clean,'start_day','duration_min','member_gender',order=order,palette=['#0a306b','#f83194','#f4e532'])
```
